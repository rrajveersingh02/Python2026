# Set -->

- If we want to represent a group of unique values as a single entity then we should go for set.
- Duplicates are not allowed.
- Insertion order is not preserved.But we can sort the elements.
- Indexing and slicing not allowed for the set.
- Heterogeneous elements are allowed.
- Set objects are mutable i.e once we creates set object we can perform any changes in that object based on our requirement.
- We can represent set elements within curly braces and with comma seperation
- We can apply mathematical operations like union, intersection, difference etc on set objects.

```python
s={10,20,30}
type(s)
```

```python
s1 = {10, 20, 30, 40}         # union
s2 = {100, 200, 300, 400}
print(s1.union(s2))

print(s1 | s2)
```

```python
Days1 = {"Monday","Tuesday","Wednesday","Thursday", "Sunday"}
Days2 = {"Friday","Saturday","Sunday"}

print(Days1|Days2)

print(Days1.union(Days2))
```

```python
s1 = {10, 20, 300, 40}
s2 = {100, 200, 300, 400}

print(s1.intersection(s2))

print(s1 & s2)
```

```python
Days1 = {"Monday","Tuesday", "Wednesday", "Thursday"}
Days2 = {"Monday","Tuesday","Sunday", "Friday"}

print(Days1&Days2) #prints the intersection of the two sets

print(Days1.intersection(Days2))
```

```python
s1 = {10, 20, 300, 40}
s2 = {100, 200, 300, 400}

            # difference(): x.difference(y) OR x-y, Returns the elements present in x but not in y

print(s1.difference(s2))
print(s1 - s2)
```

```python
s1={10,20,300,40}
s2={100,200,300,400}

               # Returns elements present in either x OR y but not in both.

print(s1.symmetric_difference(s2))
print(s1^s2)
```

# frozenset -->represent group of elements same as set but it is immutable


```python
s={10,20,"Sample"}
fs=frozenset(s)
type(fs)
```

```python
t=(10,20,30,40)
print(type(t))
fs=frozenset(t)
print(fs)
print(type(fs))
```
