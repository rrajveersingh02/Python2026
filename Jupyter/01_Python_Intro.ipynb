# print odd/ even number using range()

```python
for i in range(1,10,2):  # odd number
  print(i)
```

```python
for i in range(2,10,2):  # even number
  print(i)
```

```python
# Square root of odd  number
for i in range(1,10,2):
  print(i**2)
```

```python
# Square root of even number
for i in range(2,10,2):
  print(i**2)
```

```python
# Square root of even number
for i in range(2,10,2):
  print(i**2, end='')
```

# Check Prime

```python
n=7
for i in range(2,n):
  if n%i==0:
    print("no is not prime")
    break
  elif n%i!=0:
    print('no is prime')
    break
  else:
    print("Invalid")
```
