# Dict -->> If we want to represent a group of objects as key-value pairs then we should go for Dictionary.

Eg:-
 rollno ---- name
 phone number -- address
 ipaddress --- domain name

- Duplicate keys are not allowed but values can be duplicated.
- Hetrogeneous objects are allowed for both key and values.
- Insertion order is not preserved
- Dictionaries are mutable
- Dictionaries are dynamic
- indexing and slicing concepts are not applicable

```python
dict1 = {
    "name": "Rajveer",
    "dist": "Indore",
    "city": "Mhow"
}
x = dict1.get("name")  # getname
print(x)
# print("Accessing a element using get:")
k = dict1.keys()  # dict keys
v = dict1.values()  # dict values
print(k)
print(v)
```

```python
d = {100: 'Raj', 200: 'Singh', 300: 'CDAC'}
print(d)
l1=list(d.keys())
l2=list(d.values())
print(l1)
print(l1[1])
print(l2)
print(l2[2])
```

```python
# Looping

for i in d:
    print(d[i])

print('#########################################')

for x in d.values():
    print(x)

print('#########################################')

for x in d.keys():
    print(x)

print('#########################################')

for x, y in d.items():
    print(x, y)
```
