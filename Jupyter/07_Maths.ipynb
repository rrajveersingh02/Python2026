# Combinations

```python
math.comb(10,4)
```

```python
def combination(n,r):
  num= math.factorial(n)
  den=math.factorial(r) * math.factorial(n-r)
  return num/den

combination(10,4)
```

# Permutation

```python
math.perm(10,4)
```

```python
def permutation(n,r):
  num=math.factorial(n)
  den=math.factorial(n-r)
  return num/den

permutation(10,4)
```
