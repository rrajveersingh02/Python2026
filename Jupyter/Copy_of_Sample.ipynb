```python
a=[12,23,43,5,1,4,89,500,11]
n=len(a)
```

```python
for i in range(n):      # Bubble sort
  for j in range(0,n-1):
    if a[j] > a[j+1]:
      a[j], a[j+1] = a[j+1],a[j]
```

```python
print(a)
```

```python
a=[12,23,43,5,1,4,89,500,11]
key=1

flag=False
for i in range(len(a)):
  if a[i] == key:
    flag=True
    break
if flag ==True:
  print('Element found at Index ', i)
else:
  print("element not found")
```
