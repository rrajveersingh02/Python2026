# Decorator -->>  
- Decorator is a function which can take a function as argument and extend its functionality and returns modified function with extended functionality.

input function ----->> Decorator ---->> Output function(with extended functionality)

```python
def f1():
  print("this is first func..")

f1()
```

```python
def decor(func):  # display() pass as argument here
    def inner():
        print("enhance functionality of display function")
    return inner


@decor  # pass decorator function name
def display():
    print("show as it is ")


display()
```

```python
def newDecor(func):
  print("this is decor func..")

def display():
  print("This is normal func..")

display()
```

```python
def newDecor(func):
  def inner():
    print("this is decor func..")
  return inner

@newDecor   # newDecor is the decorator function name
def display():
  print("This is normal func..")

display()
```

```python
def addDecor(func):
    def inner(a, b):   # parameters
        print('#' * 20)
        print("sum ", end='')
        func(a, b)                # call add function
        print('#' * 20)

    return inner


@addDecor
def add(a, b):
    print(a + b)


add(10, 20)
```

```python
def decor(func):
    def inner(name):
        if name == 'sunny':
            print("hello sunny very good morning ")
        else:
            func(name)
    return inner

@decor
def wish(name):
    print("hello",name)


wish("sunny")
```

```python
def decor(func):   #wish is func here
    def inner(name):
        names = ['CM','PM','Minister']
        if name in names:
            print("hello VIP very good morning ")
        else:
            func(name)
    return inner

@decor
def wish(name):
    print("hello",name)


wish("CM")
```

```python
def newDecor(func):
  def new(a,b):
    func(a,b)
  return new

@newDecor
def add(a,b):
  print(a+b)

add(100,20)
```

```python
def divDecor(func):
    def inner(a, b):
        if b == 0:
            print("can't divide with zero")
        else:
            func(a, b)

    return inner

@divDecor
def div(a, b):
    print(a / b)


div(10, 2)
div(10, 0)
```

```python
# how to call same function with or without decorator

def decor(func):
    def inner(name):
        if name == 'sunny':
            print("hello sunny very good morning ")
        else:
            func(name)
    return inner

def wish(name):
    print("hello",name)


decorated_wish = decor(wish)

wish('raj')              # wish function called

decorated_wish("raj")      # wish function called because name is not sunny

decorated_wish('sunny')     # decorator function called
```

```python
def decor1(func): # f() is input function for this
    def inner1():
        print("Decor 1 function...")
        func()  # call original function i.e. f()

    return inner1


def decor2(func): # inner1 will be input function for this
    def inner2():
        print("Decor 2 func ..")
        func() # call input function recurson error

    return inner2

@decor2
@decor1  # first dacor1 will run then decor 2 means nearest will run first
def f():
    print("original func...")


f()

# conclusion - f() is input for decor1 and inner1 will be output of decor1,
# similarly inner1 will be input for decor2 and inner2 will be output for decor2
```

# Generator -->>

 - it is special type of function which is responsible to generate a sequence of values, it won't take any input,generator function use internally a special keyword, which is used to return generated values, keyword called yield.

- we can write generator function just like ordinary function, but it uses special keyword 'yield' to return values.

- ** yield keyword is used in generator to generate values.

- to geneate sequence of values

- can handle huge amount of data

- generator won't store/create any object, so no memory error

- If the body of a function contains yield, the function automatically becomes a generator function.

# Advantages -->

- Memory utilization is improved

- performance improve

- handle huge amount of data

# Limitations -->

- no storage of Data


```python
g = (x*x for x in range(10))
print(type(g))  # return generator type
```

```python
def gen():
    yield 'A'
    yield 'B'
    yield 'C'


g = gen()
print(type(g))

print(next(g))
print(next(g))
print(next(g))
```

```python
def gen():
  yield "Raj"
  yield "Singh"
  yield 100

g=gen()
print(next(g))
print(next(g))
print(next(g))
```

```python
def gen():
  yield 100
  yield 200
  yield 300
  yield 'A'
  yield 20.23
  yield 20+20j

g=gen()
for i in g:
  print(i)
```

```python
def gen(n):
  i=1
  while i <= n:
    yield i
    i+=1

g=gen(10)
for i in g:
  print(i)
```

```python
import time
def countDown(n):
  while n>=1:
    yield n
    n=n-1

c=countDown(20)
for i in c:
  print(i)
  time.sleep(1)
```

```python
# generate fabonacci

def gen():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b


g = gen()
for a in g:
    if a <= 20:
        print(a)
    else:
        break
```

```python
# Convert value present in generator into list

def gen(n):
    i = 1
    while i <= n:
        yield i
        i = i + 1


g = gen(10)
l=list(g)
print(l)
```
