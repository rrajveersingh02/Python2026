# Variable-length Arguments (*args) -->>

In large projects, sometimes we may not know the number of arguments to be passed in advance. In such cases, Python provides us the flexibility to offer the comma-separated values which are internally treated as tuples at the function call. By using the variable-length arguments, we can pass any number of arguments.


- after variable length arg if we are taking other argument we provide values as keyword argument only

- if first arg is keyword/positional & second or last arg is var length arg then there is no restriction,

- restriction apply if the first arg is var len arg & compulsory second args is keyword.

- function can't take more than one var args

- internally tuple will be create

```python
def printme(*names):
    print("type of passed argument is ",type(names))
    print("printing the passed arguments...")
    for name in names:
        print(name)

printme("john","David","smith","nick")
```

# **kwargs (Keyword variable length Arguments) -->internally it is 'dict'

- type of  key value representation

- def display(**kwargs): we can call this function by passing any number of keyword argument, these keyword become dict.

- def display(**kwargx): is also valid


```python
def display(**kwargs):
	print("Info...")
	for k,v in kwargs.items():
		print(k,".....",v)

display(name="Rajveer", marks="100")
display(name="Raj", marks="200")
```

# Filter function -->>

- We can use filter() function to filter values from the given sequence based on some condition.

- filter(function,sequence) Where Function Argument is responsible to perform conditional check Sequence can be List OR Tuple OR String.

```python
def isEven(n):          # without filter
    if n % 2 == 0:
        return True
    else:
        return False

l = [1, 2, 3, 4, 5, 6, 78, 8, 9, 10]
l1 = []
for n in l:
    if isEven(n) == True:
        l1.append(n)

print(l1)
```

```python
def isEven(n):    # with filter
    if n % 2 == 0:
        return True
    else:
        return False

l = [0, 10, 20, 13, 26, 22, 19]
l1 = list(filter(isEven, l))
print(l1)
```

```python
l = [0, 10, 20, 13, 26, 22, 19]            # filter with lambda
l1 = list(filter(lambda x: x % 2 == 0, l))
print(l1)
```

```python
l = [0, 3, 9, 10, 20, 13, 26, 22, 19]
l1 = list(filter(lambda x: x % 3 == 0 and x % 2 != 0, l))
print(l1)

l = ["Raj", "panky", "Viv"]
l1 = list(filter(lambda x: x[0] == 'R', l))
print(l1)

l = ["Raj", "panky", "Viv"]
l1 = list(filter(lambda x: len(x) % 3 == 0, l))
print(l1)
```

# Map function -->>

- map() used to apply function on all the elements of sequence

- for every value mapped value will generate, if input contain 5 element output should contain 5 element

```python
def double(x):
    return 2 * x


l = [2, 3, 5, 76, 3]
l1 = list(map(double, l))
print(l1)
```

# Reduce -->>

- reduce() function reduces sequence of elements into a single element by applying the specified function

- reduce(function,sequence)
- reduce() function present in functools module and hence we should write import statement.

```python
from functools import reduce

l = [10, 20, 30, 40, 50]
result = reduce(lambda x, y: x + y, l)
print(result)
```

```python
# reduce
from functools import reduce

result = reduce(lambda x, y: x + y, range(1, 101))
print(result)
```
